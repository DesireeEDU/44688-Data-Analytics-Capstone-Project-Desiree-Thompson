### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

### Import Data Source

```python
df = pd.read_csv('bodyfat.csv')
```

### Data Description
- The data set has 15 features.

- There are no missing values.

```python
df['Height (m)'] = df['Height'] * 0.0254 # 1 inch = 0.0254 meters
df['Weight (kg)'] = df['Weight'] * 0.453592 # 1 pound = 0.453592 kilograms
df['BMI'] = df['Weight (kg)']/(df['Height (m)'] ** 2)
df['BMI'] = df['BMI'].round(2)
df.head(5)
```

```python
# Calculate the Body Fat Percentage using the given formula
df['BMI_Body_Fat'] = 1.20 * df['BMI'] + 0.23 * df['Age'] - 16.2

# Limit the output to two decimal points
df['BMI_Body_Fat'] = df['BMI_Body_Fat'].round(2)

# Save the updated DataFrame back to a CSV file
df.to_csv('bodyfat.csv', index=False)
df.head(10)
```

### 

```python
# Convert height and abdomen from centimeters to inches (1 cm = 0.393701 inches)
df['neck_in'] = df['Neck'] * 0.393701
df['abdomen_in'] = df['Abdomen'] * 0.393701


# Calculate the Body Fat Percentage using the given formula
df['Navy_Body_Fat'] = (
    86.010 * np.log10(df['abdomen_in'] - df['neck_in']) -
    70.041 * np.log10(df['Height']) +
    36.76
)

# Limit the output to two decimal points
df['Navy_Body_Fat'] = df['Navy_Body_Fat'].round(2)

# Save the updated DataFrame back to a CSV file
df.to_csv('bodyfat.csv', index=False)
df.head(10)
```

```python
# List of columns to remove
columns_to_remove = ['BMI Body Fat', 'Height (m)', 'Weight (kg), 'neck_in', 'abdomen_in', 'Navy_Body_Fat', 'BMI_Body_Fat_Percentage', 'BMI_Body_Fat']

# Drop the specified columns
df = df.drop(columns=columns_to_remove)

# Save the updated DataFrame back to a CSV file
df.to_csv('bodyfat.csv', index=False)
```
